```python
# bibliotecas
import os
import pandas as pd
from pathlib import Path
```

```python
# Perguntar ao usuário se são dados de desenvolvimento ou produção
data_type = input("Você está trabalhando com dados de desenvolvimento ou produção? (Digite '1' para 'dev' ou '2' para 'prod'): ").strip().lower()
```

```python
# Verificar o tipo de dados e definir os caminhos correspondentes
if data_type == '1':
    file_name_intermediate = 'data_filtered_dev.parquet'
    feature_suffix = 'dev'
elif data_type == '2':
    file_name_intermediate = 'data_filtered_prod.parquet'
    feature_suffix = 'prod'
else:
    raise ValueError("Tipo de dados inválido. Por favor, digite 'dev' ou 'prod'.")
```

```python
# Diretório base do projeto
BASE_DIR = Path(os.getcwd()).parent

# Caminho onde os dados processados (intermediários) estão armazenados
PRIMARY_DATA_PATH = BASE_DIR / 'data' / '02_intermediate'

# Caminho para salvar os dados com features extraídas
FEATURE_DATA_PATH = BASE_DIR / 'data' / '04_feature'
os.makedirs(FEATURE_DATA_PATH, exist_ok=True)
```

```python
# Caminho do arquivo de entrada
INPUT_FILE_PATH = PRIMARY_DATA_PATH / file_name_intermediate

# Caminhos dos arquivos de saída
X_FEATURES_FILE_PATH = FEATURE_DATA_PATH / f'x_features_{feature_suffix}.parquet'
Y_FEATURES_FILE_PATH = FEATURE_DATA_PATH / f'y_features_{feature_suffix}.parquet'
```

```python
# Ler o arquivo correspondente
data = pd.read_parquet(INPUT_FILE_PATH)
```

```python
data.columns
```

```python
# Selecionar as colunas que serão usadas para a extração de features
features = ['lat', 'lon', 'minutes_remaining', 'period', 'playoffs', 'shot_distance', 'shot_made_flag']

# Separar X e y usando a variável 'features'
X = data[features[:-1]]  # Variáveis independentes (todas exceto a última)
y = data[[features[-1]]]  # Variável dependente (última coluna)
```

```python
# Salvar os arquivos de X e y
X.to_parquet(X_FEATURES_FILE_PATH, index=False)
y.to_parquet(Y_FEATURES_FILE_PATH, index=False)

print(f"Arquivos de features salvos em:\n{X_FEATURES_FILE_PATH}\n{Y_FEATURES_FILE_PATH}")
```
